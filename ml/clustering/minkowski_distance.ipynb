```python
# Código para calcular y visualizar las distancias de Minkowski
import numpy as np
import matplotlib.pyplot as plt

# Vectores de ejemplo
x1 = np.array([1, 2])
x2 = np.array([4, 6])


# Calcular distancias con diferentes valores de p
def minkowski_distance(x1, x2, p):
    return np.sum(np.abs(x1 - x2) ** p) ** (1 / p)


p_values = [1, 2, 3, 10, np.inf]
distances = [minkowski_distance(x1, x2, p) for p in p_values]

# Mostrar los resultados
dist_labels = ['Manhattan (p=1)', 'Euclidean (p=2)', 'p=3', 'p=10', 'Chebyshev (p=∞)']
for label, dist in zip(dist_labels, distances):
    print(f"{label}: {dist:.4f}")

# Graficar los puntos y las líneas
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot([x1[0], x2[0]], [x1[1], x2[1]], 'k--', label='Vector line')
ax.plot(x1[0], x1[1], 'ro', label='x1 (1,2)')
ax.plot(x2[0], x2[1], 'bo', label='x2 (4,6)')
ax.set_xlim(0, 7)
ax.set_ylim(0, 8)
ax.set_title('Distancias Minkowski entre x1 y x2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True)
ax.legend()
plt.show()
```
